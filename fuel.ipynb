# Fuel Price Predictor

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv("https://raw.githubusercontent.com/Saurabh-Datta/Fuel_Price_Predictor/main/petrol.csv")
```

```
X = dataset.iloc[:, 0].values
y = dataset.iloc[:, 1:].values
```

```
print(y)
```

```
X.size
```

```
m = [0]*374
```

```
mo = [0]*374
```

```
i=0
while (i<374):
  if (X[i].find("January") != -1):
    m[i]=0
    mo[i]=1
  if (X[i].find("February") != -1):
    m[i]=31
    mo[i]=2
  if (X[i].find("March") != -1):
    m[i]=59
    mo[i]=3
  if (X[i].find("April") != -1):
    m[i]=90
    mo[i]=4
  if (X[i].find("May") != -1):
    m[i]=120
    mo[i]=5
  if (X[i].find("June") != -1):
    m[i]=151
    mo[i]=6
  if (X[i].find("July") != -1):
    m[i]=181
    mo[i]=7
  if (X[i].find("August") != -1):
    m[i]=212
    mo[i]=8
  if (X[i].find("September") != -1):
    m[i]=243
    mo[i]=9
  if (X[i].find("October") != -1):
    m[i]=273
    mo[i]=10
  if (X[i].find("November") != -1):
    m[i]=304
    mo[i]=11
  if (X[i].find("December") != -1):
    m[i]=334
    mo[i]=12
  i+=1
```

```
m
```

```
ye = [0]*374
```

```
i = 0
while (i<374):
  ye[i]= (int(X[i][-4:])-2013)
  i+=1
```

```
ye
```

```
i = 0
date = [0]*374
while (i<374):
  if (ye[i]==1):
    date[i] = 0
  if (ye[i]==2):
    date[i] = 365
  if (ye[i]==3):
    date[i] = 730
  if (ye[i]==4):
    date[i] = 1096
  if (ye[i]==5):
    date[i] = 1461
  date[i] += m[i]
  if (ye[i]==3 and mo[i]>2):
    date[i] +=1
  date[i] += int(X[i][-7:-5])
  i+=1
```

```
date[0]
```

```
X = date
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training the Random Forest Regression model on the whole dataset

```
X = np.array(X)
```

```
X.reshape(374,1)
```

```
X= X.reshape(374,1)
```

```
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X_train, y_train)
```

## Predicting the Test set results

```
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
```

```
y_pred
```

```
y_test
```

## Evaluating the Model Performance

```
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
